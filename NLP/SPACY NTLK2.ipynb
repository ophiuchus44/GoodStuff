```python
def stem(tokens):
    '''rules-based stemming of a bunch of tokens'''
    
    new_bag = []
    for token in tokens:
        # define rules here
        if token.endswith('ing'):
            new_bag.append(token[:-3])
        
    return new_bag

len(stem(text1))
```
